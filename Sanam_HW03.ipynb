```python
# Question1

import matplotlib.pyplot as plt
import matplotlib.patches as patches

class rectangle:                                        #a.creating a class rectangle with 4 arguments (length, width, x and y coordinates)
  def __init__(self, a, b, x1, y1):                            
    self.a= a
    self.b= b
    self.x1= x1
    self.y1= y1

  def area(self):                                       #b.creating a method to calculate area
    area= self.a * self.b
    return area

  def perimeter(self):                                  #c.creating a method to calculate perimeter
    perimeter = 2*(self.a+ self.b)
    return perimeter

  def plot_shape(self):                                 #d.plotting the rectangle using the co-ordinates and dimensions
    fig1 = plt.figure()
    ax1 = fig1.add_subplot(111, aspect='equal')
    ax1.add_patch(patches.Rectangle((self.x1, self.y1), self.a, self.b, facecolor='none',edgecolor= 'blue',linewidth= 2))
    plt.ylim((0, 10))
    plt.xlim((0, 10))
    plt.show()
    

  def shift(self, delta_x, delta_y):                     #e.shifing the rectangle 
    self.x1 += delta_x
    self.y1 += delta_y

```

```python
# create a new rectangle object
r1 = rectangle(5,3,0,0)
```

```python
# get the area
r1.area()
```

```python
# get the perimeter
r1.perimeter()
```

```python
# plot the rectangle
r1.plot_shape()
```

```python
# shift the rectangle
r1.shift(2,2)
```

```python
# plot it and check the shift
r1.plot_shape()
```

```python
# Question2 

class square(rectangle):                               #creating a class square by inheriting from the rectangle class
  def __init__(self, a,x1,y1):
    super(square, self).__init__(a, a,x1,y1)           #overiding the __init__ method 
```

```python
# create a square

s1 = square(4,2,2)
```

```python
# area test
s1.area()
```

```python
# perimeter test
s1.perimeter()
```

```python
# plot
s1.plot_shape()
```
